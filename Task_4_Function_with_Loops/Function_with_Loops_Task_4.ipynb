Problem 1: Mobile Recharge Validation System

```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return False

    if amount not in valid_plans:
        return False

    return True


# Retry using while loop
while True:
    amount = int(input("Enter recharge amount: "))

    if validate_recharge(amount):
        print("Recharge Successful")
        break
    else:
        print("Invalid Recharge Amount Please retry.\n")
```

Problem 2: Inventory Reorder Alert System

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(stock_dict):
    for product, stock in stock_dict.items():
        if stock < 15:
            print(product, "→ Reorder Alert ")
        else:
            print(product, "→ Stock OK")


# Sample Dictionary
products = {
    "Rice Bags": 10,
    "Sugar Packs": 20,
    "Oil Bottles": 12,
    "Soap": 25
}

check_inventory(products)
```

Problem 3: Student Result Processing System

```python
# Problem 3: Student Result Processing System

def student_result(marks):
    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Example
marks_list = [45, 60, 55]
result = student_result(marks_list)
print("Result:", result)
```

Problem 4: Cab Fare Estimator with Retry Option

```python
# Problem 4: Cab Fare Estimator with Retry Option

def calculate_fare(km, peak):
    base_fare = 50
    total = base_fare + (km * 12)

    if peak == "yes":
        total += total * 0.25   # 25% extra

    return total


while True:
    km = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    fare = calculate_fare(km, peak)
    print("Total Fare = ₹", fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

Problem 5: Employee Attendance Eligibility Checker

```python
# Problem 5: Employee Attendance Eligibility Checker

def attendance_checker(attendance_list):
    present = 0

    for day in attendance_list:
        if day == "P":
            present += 1

    percentage = (present / len(attendance_list)) * 100
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example
attendance = ["P", "P", "A", "P", "P"]
status = attendance_checker(attendance)
print("Status:", status)
```

Problem 6: Password Strength Checker

```python
# Problem 6: Password Strength Checker

def password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password (Minimum length should be 8)"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special char)"


# Input Password
pwd = input("Enter password: ")
print(password_strength(pwd))
```
