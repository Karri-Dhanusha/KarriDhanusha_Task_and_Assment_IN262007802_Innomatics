Problem 1: User Login Check

```python
# Predefined credentials
username = "admin"
password = "1234"

# Taking user input
input_username = input("Enter username: ")
input_password = input("Enter password: ")

# Checking credentials
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

Problem 2: Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students Passed:", pass_count)
print("Total Students Failed:", fail_count)
```

Problem 3: Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("Cleaned Names:", cleaned_names)
```

Problem 4: Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")

    if length > 10:
        print("-> This message is longer than 10 characters.")
```

Problem 5: Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR messages:", error_count)
```
