Problem 1: Unique Words in a Sentence


```python
# Unique Words in a Sentence

# Input from user
sentence = input("Enter sentence: ")

# Split sentence into words
words = sentence.split()

# Convert list to set to get unique words
unique_words = set(words)

# Print count of unique words
print("Unique words count:", len(unique_words))

# Print unique words set
print("Unique words:", unique_words)
```

Problem 2: Highest Salary from Employee Data

```python
# Highest Salary from Employee Data

# Employee dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)

# Print result
print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

Problem 3: Find Maximum and Minimum Values

```python
# Find Maximum and Minimum Values

# Input list
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with the first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through the list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Print results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

Problem 4: Count Products Above a Price Threshold

```python
# Count Products Above a Price Threshold

# Input list of product prices
prices = [450, 1200, 899, 1500, 300]

# Initialize count
count = 0

# Loop through prices
for price in prices:
    if price > 1000:
        count += 1

# Print result
print("Products above 1000:", count)
```

Problem 5: Calculate Attendance Percentage

```python
# Calculate Attendance Percentage

# Input attendance list
attendance = ["P", "P", "A", "P", "P"]

# Count number of Present days
present_count = attendance.count("P")

# Total number of days
total_days = len(attendance)

# Calculate attendance percentage
percentage = (present_count / total_days) * 100

# Print result
print("Attendance Percentage:", percentage)
```

Problem 7: Count Character Frequency

```python
# Count Character Frequency

# Input string
text = "pythonp"

# Dictionary to store character counts
char_frequency = {}

# Loop through each character
for char in text:
    if char in char_frequency:
        char_frequency[char] += 1
    else:
        char_frequency[char] = 1

# Print result
print(char_frequency)
```

Problem 9: Check if a Key Exists in a Dictionary

```python
# Check if a Key Exists in a Dictionary

# Dictionary of employees
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Key to check
key = "Ravi"

# Check if key exists
if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

Problem 10: Calculate Average Marks

```python
# Calculate Average Marks

# List of marks
marks = [70, 75, 80, 65]

# Calculate sum of marks
total = sum(marks)

# Calculate number of subjects
count = len(marks)

# Calculate average
average = total / count

# Print result
print("Average Marks:", average)
```
